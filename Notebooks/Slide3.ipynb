```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gt_df = pd.read_csv('/Users/jxhn/Desktop/IRONHACK/Week_5/Mid_bootcamp_project/Data/Raw/globalterrorismr.csv', encoding = 'ISO-8859-1')
```

```python
attacks_per_gang = gt_df.groupby('gname').size().reset_index(name='count').sort_values(by='count', ascending = False)
top10_gangs = attacks_per_gang.head(10)
```

```python
filtered_gangs = top10_gangs[top10_gangs['gname'] != 'Unknown'] #removing 'Unknown' column
```

```python
plt.figure(figsize=(10, 6))  # Adjusted the figure size as needed
ax = sns.barplot(x='gname', y='count', data=filtered_gangs)
ax.set_title('Top 10 Gangs with the highest number of terrorist attacks between 1970 to 2017')
ax.set_xlabel('Gangs')
ax.set_ylabel('Number of Terrorist Attacks')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')  # Rotate x-axis labels

plt.tight_layout()  # Adjust the layout of x-axis 
plt.show()
```

```python
kills_per_gang = gt_df.groupby(['gname', 'success']).size().reset_index(name='count').sort_values(by='count', ascending = False)
top10_gangkills = kills_per_gang.head(10)
```

```python
filtered_kills = top10_gangkills[top10_gangkills['gname'] != 'Unknown'] #removing 'Unknown' column
```

```python
plt.figure(figsize=(10, 6))  # Adjusted the figure size as needed
ax = sns.barplot(x='gname', y='count', data=filtered_kills)
ax.set_title('Top 10 Gangs with highest number of successful terrorist attacks between 1970 to 2017')
ax.set_xlabel('Gang')
ax.set_ylabel('Number of successful terrorist attacks')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')  # Rotate x-axis labels

plt.tight_layout()  # Adjust the layout of x-axis 
plt.show()
```

```python
attacktype_per_gang2 = gt_df.groupby(['gname', 'success']).size().reset_index(name='count')
success_per_attacktype = attacktype_per_gang2[attacktype_per_gang2['success'] == 1]
top10_attacktypes2 = success_per_attacktype.groupby('gname')['count'].sum().nlargest(10).reset_index(name='success_count')

top10_attacktypes2['total_count'] = attacktype_per_gang2.groupby('gname')['count'].sum()

# Filter out 'Unknown' category
top10_attacktypes2 = top10_attacktypes2[top10_attacktypes2['gname'] != 'Unknown']

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='gname', y='total_count', data=top10_attacktypes2, color='blue', label='Total Attacks')
sns.barplot(x='gname', y='success_count', data=top10_attacktypes2, color='green', alpha=0.7, label='Successes')

ax.set_title('Top 10 Gangs with Highest Success Rate (at least one person was killed) of Terrorist Attacks between 1970 and 2017')
ax.set_xlabel('Gang')
ax.set_ylabel('Number of Attacks')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```
